```python
# Handle categorical variables
df = pd.get_dummies(df, columns=["Sex", "Education"])

# select features and target
X = df.drop("HeartDiseaseorAttack", axis=1)
y = df["HeartDiseaseorAttack"]
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler
```

```python
# Perform oversampling
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)
```

```python
# Perform feature selection
selector = SelectKBest(score_func=chi2, k=10)
X_new = selector.fit_transform(X_resampled, y_resampled)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_new, y_resampled, test_size=0.2, random_state=42)
```

```python
# Scale the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Define the models
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
```

```python
models = {'Logistic Regression': LogisticRegression(random_state=42),
          'Ridge Classifier' : RidgeClassifier(),
          'Random Forest': RandomForestClassifier(random_state=42),
          'Gradient Boosting': GradientBoostingClassifier(random_state=42),
          'Naive Bayes': GaussianNB(),
          'KNN': KNeighborsClassifier(),
          'Decision Tree': DecisionTreeClassifier(random_state=42),
          'Neural Network': MLPClassifier(random_state=42)}    
```

```python
# Perform oversampling
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)

# Perform feature selection
from sklearn.feature_selection import SelectKBest
k = 10
selector = SelectKBest(k=k)
X_resampled = selector.fit_transform(X_resampled, y_resampled)

kf = KFold(n_splits=5)
for model_name, model in models.items():
    cv_scores = []
    for train_index, test_index in kf.split(X_resampled):
        X_cv_train, X_cv_test = X_resampled[train_index], X_resampled[test_index]
        y_cv_train, y_cv_test = y_resampled[train_index], y_resampled[test_index]
        model.fit(X_cv_train, y_cv_train)
        cv_scores.append(model.score(X_cv_test, y_cv_test))
    print(f"{model_name} - Cross Validation Score: {np.mean(cv_scores)} +/- {np.std(cv_scores)}")
    
   # Fit the model and make predictions on the test set
    model.fit(X_resampled, y_resampled)
    y_pred = model.predict(X_test)

    # Evaluate the model
    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
    print(f"{model_name} - Test Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print(f"{model_name} - Test Precision: {precision_score(y_test, y_pred)}")
    print(f"{model_name} - Test Recall: {recall_score(y_test, y_pred)}")
    print(f"{model_name} - Test F1-Score: {f1_score(y_test, y_pred)}")

    
```
